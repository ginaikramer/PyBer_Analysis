```python
import matplotlib
matplotlib.__version__
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
savings = [0,100,150,200,300,500,300,350,500, 600,750,1000]
months = ['Jan', 'Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
plt.plot(months,savings)
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Savings Account Results')
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="red", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
#Add a grid
plt.grid()
# Add the legend.
plt.legend()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis,marker="*", color="green", linewidth=2, label="Boston")
ax.set_xlabel("Date")
ax.set_ylabel("Fares ($)")
ax.set_title("PyBer Fares By Month")
ax.set_ylim(0,45)
ax.grid()
```

```python
plt.bar(x_axis,y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#Create a horizontal bar chart
#plt.barh(x_axis,y_axis)
#plt.gca().invert_yaxis()

#Skill drill
plt.barh(x_axis,y_axis,color="magenta",label="Boston")
plt.gca().invert_yaxis()
plt.ylabel("Date")
plt.xlabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#Create a vertical bar chart using object oriented methods
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
#Create a horizontal bar chart using object oriented methods
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
#Create a horizontal bar chart and switch axes
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.set_xlabel("Fares ($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fares by Month")
ax.invert_yaxis()
ax.legend()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
#Bubble chart
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
#scatter plot using the object oriented way
fig, ax=plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
#bubble chart using the object oriented way
fig, ax=plt.subplots()
ax.scatter(y_axis,x_axis, s=[i*5 for i in y_axis], color="blue", label="Boston")
ax.invert_yaxis()
ax.set_xlabel("Fares ($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fares by Month")
ax.legend()
```

```python
#Create a pie chart
explode_values=(0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, labels=x_axis, explode=explode_values,autopct='%.1f%%')
plt.show()
```

```python
#Create another pie chart with different attributes
explode_values=(0,0,0,0,0,0,0.2,0,0,0,0,0)
colors=["slateblue","magenta","lightblue","green","yellowgreen","blue","yellow", "orange","gold","purple", "tomato","mistyrose"]
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
        labels=x_axis, 
        colors=colors,
        explode=explode_values,
        autopct='%.1f%%')
plt.show()
```

```python
#Create pie chart using object oriented
fig, ax=plt.subplots(figsize=(8,8))
explode_values=(0,0,0.2,0,0,0,0.2,0,0,0,0,0)
colors=["slateblue","magenta","lightblue","green","yellowgreen","blue","yellow", "orange","gold","purple", "tomato","mistyrose"]
ax.pie(y_axis, labels=x_axis,explode=explode_values, colors=colors, autopct="%.1f%%", startangle=90)
plt.show()
```

```python
import statistics
```

```python
stdev = statistics.stdev(y_axis)
stdev
```

```python
stdev2 = statistics.stdev([1,5,9,13,17])
stdev2
```

```python
#Add errorbard using Matlab
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
#Add errorbars using object oriented 
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
#Add error bars and caps to bar chart using matlab
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
#Add error bars and caps to bar chart using OO
fig, ax=plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
#use xticks to spread x axis out in small increments using matlab
import numpy as np
plt.barh(x_axis,y_axis)
plt.xticks(np.arange(0,51,step=5.0))
plt.gca().invert_yaxis()
```

```python
#use xticks to spread x axis out in small increments using OO
fig, ax=plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0,51,step=5.0))
#or
#ax.xaxis.set_ticks(np.arange(0,51,step=5.0))
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator
fig, ax=plt.subplots(figsize=(8,8))
ax.barh(x_axis, y_axis)
ax.xaxis.set_ticks(np.arange(0,51,step=5.0))
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
